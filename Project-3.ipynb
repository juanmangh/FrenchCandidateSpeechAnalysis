## 2. Word2Vec

```python
# Download the punkt tokenizer for sentence splitting
import nltk.data
#nltk.download()   

# Load the punkt tokenizer
tokenizer = nltk.data.load('tokenizers/punkt/french.pickle')

# Define a function to split a review into parsed sentences
def speech_to_sentences( review, tokenizer, remove_stopwords=True ):
    # Function to split a review into parsed sentences. Returns a 
    # list of sentences, where each sentence is a list of words
    #
    # 1. Use the NLTK tokenizer to split the paragraph into sentences
    raw_sentences = tokenizer.tokenize(review.strip())
    #
    # 2. Loop over each sentence
    sentences = []
    for raw_sentence in raw_sentences:
        # If a sentence is empty, skip it
        if len(raw_sentence) > 0:
            # Otherwise, call review_to_wordlist to get a list of words
            sentences.append( speech_to_wordlist( raw_sentence, \
              remove_stopwords=True ))
    #
    # Return the list of sentences (each sentence is a list of words,
    # so this returns a list of lists
    return sentences
```

# 3. Function

```python
def word2vec_mine( sentences, name ):
    # Import the built-in logging module and configure it so that Word2Vec 
    # creates nice output messages
    import logging
    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',\
        level=logging.INFO)

    # Set values for various parameters
    num_features = 30    # Word vector dimensionality                      
    min_word_count = 10   # Minimum word count                        
    num_workers = 4       # Number of threads to run in parallel
    context = 5          # Context window size                                                                                    
    downsampling = 1e-3   # Downsample setting for frequent words

    # Initialize and train the model (this will take some time)
    from gensim.models import word2vec
    print ("Training model...")
    #model = word2vec.Word2Vec(lepen_tok_sentences_list, workers=num_workers, \
    #            size=num_features, min_count = min_word_count, \
    #            window = context, sample = downsampling)
    model = word2vec.Word2Vec(sentences=sentences, size=100, alpha=0.1, window=5, \
                              min_count=10, \
                              workers=5, sg=0, hs=0, negative=0, cbow_mean=0, \
                              iter=10, sorted_vocab=1)
    # If you don't plan to train the model any further, calling 
    # init_sims will make the model much more memory-efficient.
    model.init_sims(replace=True)

    # It can be helpful to create a meaningful model name and 
    # save the model for later use. You can load it later using Word2Vec.load()
    model_name = name
    model.save(model_name)
    
    
```

# Create matrix for Sequential

```python
def create_matrix(name,matrixname):
    model = word2vec.Word2Vec.load(name)
    w, h = 100, len(model.wv.vocab);
    print(len(model.wv.vocab))
    a = list(model.wv.vocab.keys())
    matrixname = [[0 for x in range(w)] for y in range(h)] 
    numb = 0
    for item in a:
        for x in range(0, w):
            matrixname[numb][x] = model[item][x]
        numb = numb +1
    return matrixname;

def create_matrix_test(name,testname,matrixname,matrixtestname):
    model = word2vec.Word2Vec.load(testname)
    model1 = word2vec.Word2Vec.load(name)
    w, h = 100, len(model1.wv.vocab);
    print(len(model.wv.vocab))
    a = list(model.wv.vocab.keys())
    b = list(model1.wv.vocab.keys())
    matrixtestname = [[0 for x in range(w)] for y in range(h)] 
    numb = 0
    for item in b:
        for x in range(0, w):
            if(item in model):
                matrixtestname[numb][x] = model[item][x]
            else:
                matrixtestname[numb][x] = 0
        numb = numb +1
    return matrixtestname
```

```python
matrix_lepen = create_matrix("lepen","matrix_lepen")
matrix_lepen_test = create_matrix_test("lepen","lepen_test","matrix_lepen","matrix_lepen_test")
matrix_melenchon = create_matrix("melenchon","matrix_melenchon")
matrix_melenchon_test = create_matrix_test("melenchon","melenchon_test","matrix_melenchon","matrix_melenchon_test")
matrix_fillon = create_matrix("fillon","matrix_fillon")
matrix_fillon_test = create_matrix_test("fillon","fillon_test","matrix_fillon","matrix_fillon_test")
matrix_macron = create_matrix("macron","matrix_macron")
matrix_macron_test = create_matrix_test("macron","macron_test","matrix_macron","matrix_macron_test")
matrix_hamon = create_matrix("hamon","matrix_hamon")
matrix_jadot = create_matrix("jadot","matrix_jadot")

```

# Putting matrixes together

```python
train = np.concatenate((matrix_lepen, matrix_melenchon, matrix_fillon, matrix_macron, matrix_hamon, matrix_jadot), axis=0)
train.shape
```

```python
test = np.concatenate((matrix_lepen_test, matrix_melenchon_test, matrix_fillon_test, matrix_macron_test), axis=0)
test.shape
```

Label

```python
labelTrain1 = [1 for x in range(len(matrix_lepen))]
labelTrain2 = [2 for x in range(len(matrix_melenchon))]
labelTrain3 = [3 for x in range(len(matrix_fillon))]
labelTrain4 = [4 for x in range(len(matrix_macron))]
labelTrain5 = [5 for x in range(len(matrix_hamon))]
labelTrain6 = [6 for x in range(len(matrix_jadot))]
labelTrain = np.concatenate((labelTrain1, labelTrain2, labelTrain3, labelTrain4, labelTrain5, labelTrain6), axis=0)
print(labelTrain.shape)
```

```python
labelTest1 = [1 for x in range(len(matrix_lepen))]
labelTest2 = [2 for x in range(len(matrix_melenchon))]
labelTest3 = [3 for x in range(len(matrix_fillon))]
labelTest4 = [4 for x in range(len(matrix_macron))]
labelTest = np.concatenate((labelTest1, labelTest2, labelTest3, labelTest4), axis=0)
print(labelTrain.shape)
```

```python
from keras.datasets import reuters
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.utils import np_utils
from keras.preprocessing.text import Tokenizer
#creating the model 
model_seq = Sequential()
model_seq.add(Dense(10,input_shape=(100,) ))  #why it works with at least 5 neurons in the case of whole speeches?
model_seq.add(Activation('relu'))
model_seq.add(Dense(8))    #minimum is five even for the second layer
model_seq.add(Activation("sigmoid"))
model_seq.add(Dense(8))    #minimum is five even for the second layer
model_seq.add(Activation("softmax"))
#why it works with at least 5 neurons (Dense(5)) in the case of whole speeches?
#with this is just a Neural Network? How to make it a CNN?
```

```python
model_seq.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=["accuracy"])
```

```python
model_seq.fit(train,labelTrain,epochs=100,batch_size=1,verbose=2)
```
